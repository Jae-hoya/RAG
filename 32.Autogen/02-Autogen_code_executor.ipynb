```python
import warnings
from dotenv import load_dotenv

warnings.filterwarnings('ignore')
load_dotenv()
```

```python
import autogen
from autogen.coding import LocalCommandLineCodeExecutor

config_list = [{"model": "gpt-4o-mini"}]
```

```python
assistant = autogen.AssistantAgent(
    name="assistnant",
    llm_config={
        "config_list":config_list,
        "temperature":0
    }
)
```

```python
user_proxy = autogen.UserProxyAgent(
    name="user_proxy",
    #max_consecutive_auto_reply=10, # human_input_mode을 never로 했을때, 무한정으로 루프에 돌지 않도록, 자동응답을 10번으로 제한하는 방법.
    is_termination_msg=lambda x:x.get("content", "").rstrip().endswith("TERMINATE"), # 함수를 선언해줄건데, content에 해당하는 value값이 terminate로 끝나면 user_proxy의 행동을 멈추겠다.
    code_execution_config={
        "executor": LocalCommandLineCodeExecutor(work_dir="coding") # executor의 key값에 LocalCommandLineCodeExecutor을 넣고, coding폴더에다가 코드의 결과물을 저장하겠다.
    },
    human_input_mode="NEVER"
)
```

```python
chat_res = user_proxy.initiate_chat(
    assistant,
    message="""100까지 소수가 어떤것이 있는지 코드로 살펴봐주세요""",
    summary_method="reflection_with_llm"
)
```

```python
print("chat history:", chat_res.chat_history)
print("-"*50)
print("summary:", chat_res.summary)
print("-"*50)
print("chat info:", chat_res.cost)
```
