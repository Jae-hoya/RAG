```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from autogen import ConversableAgent

coder_agent = ConversableAgent(
    name = "Junior_Corder_Agent",
    system_message="""
    당신은 3년차 파이썬 코드 전문 소프트웨어 엔지니어 입니다.
    이해가 안되는 부분은 Senior Corder Agent에게 질문하세요.
    """,
    llm_config={"config_list": [{"model": "gpt-4o-mini"}]}
)
```

```python
manager_agent = ConversableAgent(
    name = "Senior_Corder_Agent",
    system_message="""
    당신은 20년차 파이썬 코드 전문 소프트웨어 엔지니어 입니다.
    질문이 주어지면, 해당 질문에 전문성을 가지고 대답해주세요.
    만약 코드가 주어진다면, 주어진 코드를 검토하고 효율성을 높일 방안을 탐구하세요.
    """,
    llm_config={"config_list": [{"model": "gpt-4o-mini"}]}
)
```

```python
chat_result = coder_agent.initiate_chat(
    manager_agent,
    message="피보나치 수열 코드에 대해 설명해 주세요",
    summary_method="reflection_with_llm",
    max_turns=2
)
```

```python
print(chat_result.summary)
```

# 순차적 대화방식


```python
topic_agent = ConversableAgent(
    name = "topic_agent",
    system_message=""" 당신은 토론할 주제를 제시하는 역할입니다. 현재 AI분야에서 중요한 주제를 하나 선정하여 제시해 주세요.
    """,
    llm_config={"config_list": [{"model": "gpt-4o-mini"}]},
    human_input_mode="NEVER" # 대화를 함에 있어서 사람의 개입없이 대화를 하게된다.
)
```

```python
skilled_agent = ConversableAgent(
    name = "skilled_agent",
    system_message=""" 당신은 AI 기술 전문가 역할입니다. 주어진 주제에 대해서 AI기술의 관점에서의 장단점과 전망에 대한 의견을 제시해 주세요.
    """,
    llm_config={"config_list": [{"model": "gpt-4o-mini"}]},
    human_input_mode="NEVER"
)
```

```python
economic_agent = ConversableAgent(
    name = "economic_agent",
    system_message=""" 당신은 경제학자 입니다. 주어진 주제에 대해 이전 의견들을 고려하여 경제적 관점에서 의견을 제시해 주세요.
    """,
    llm_config={"config_list": [{"model": "gpt-4o-mini"}]},
    human_input_mode="NEVER" 
)
```

```python
social_agent = ConversableAgent(
    name = "social_agent",
    system_message=""" 당신은 사회학자 입니다. 주어진 주제에 대해 이전 의견들을 고려하여 사회적 관점에서 의견을 제시해 주세요.
    """,
    llm_config={"config_list": [{"model": "gpt-4o-mini"}]},
    human_input_mode="NEVER" 
)
```

```python
environmental_agent = ConversableAgent(
    name = "environmental_agent",
    system_message=""" 당신은 환경 전문가 입니다. 주어진 주제에 대해 이전 의견들을 고려하여 환경적 관점에서 의견을 제시해 주세요.
    """,
    llm_config={"config_list": [{"model": "gpt-4o-mini"}]},
    human_input_mode="NEVER" 
)
```

```python
ethical_agent = ConversableAgent(
    name = "ethical_agent",
    system_message=""" 당신은 윤리학자 입니다. 주어진 주제에 대해 이전 의견들을 고려하여 윤리적 관점에서 의견을 제시해 주세요.
    """,
    llm_config={"config_list": [{"model": "gpt-4o-mini"}]},
    human_input_mode="NEVER" 
)
```

```python
topic = "AI 최신 기술 동향"
chat_results = topic_agent.initiate_chats(
    [
        {
            "recipient" : skilled_agent,
            "message": f"다음 주제에 대해 경제적 관점에서 의견을 제시해 주세요: {topic}",
            "max_turns":2,
            "summary_method": "last_msg"    
        },
        {
            "recipient" : economic_agent,
            "message": f"다음 주제에 대해 ai 전문가로서 의견을 제시해 주세요: {topic}",
            "max_turns":2,
            "summary_method": "last_msg"    
        },
        {
            "recipient" : social_agent,
            "message": f"다음 주제에 대해 사회적 관점에서 의견을 제시해 주세요: {topic}",
            "max_turns":2,
            "summary_method": "last_msg"    
        },
        {
            "recipient" : environmental_agent,
            "message": f"다음 주제에 대해 환경적 관점에서 의견을 제시해 주세요: {topic}",
            "max_turns":2,
            "summary_method": "last_msg"    
        },
        {
            "recipient" : ethical_agent,
            "message": f"다음 주제에 대해 윤리적 관점에서 의견을 제시해 주세요: {topic}",
            "max_turns":2,
            "summary_method": "last_msg"    
        },
    ]
)
```

```python
print("First Chat Summary:", chat_results[0].summary)
print("Second Chat Summary:", chat_results[1].summary)
print("Third Chat Summary:", chat_results[2].summary)
print("Fourth Chat Summary:", chat_results[3].summary)
```

# 그룹챗 대화 방식

```python
from autogen import GroupChat

group_chat = GroupChat(
    agents=[topic_agent, skilled_agent, economic_agent, social_agent, environmental_agent, ethical_agent],
    messages=[],
    max_round=10
)
```

```python
from  autogen import GroupChatManager

group_chat_manager = GroupChatManager(
    groupchat=group_chat,
    llm_config={"config_list": [{"model":"gpt-4o-mini"}]}
)
```

```python
chat_result = topic_agent.initiate_chat(
    group_chat_manager,
    message="AI 동향에 대해 대화를 나눠주세요",
    summary_method="reflection_with_llm"
)
```

```python
print(chat_result.summary)
```
