# OpenAI 하이퍼 파라미터
- `temperature` : 샘플린 온도. 0~2사이이며, 높은값은(ex: 0.8) 더 무작위하게 만들고(창의성), 낮은값은(ex: 0.2) 출력을 더 집중되고 결정론적으로 만든다. 1보다 큰수를 넣으면 에러가 나는 경우가 있다. 따라서 RAG를 할때는 보통 0으로 한다.
- `max_token` : 채팅 완성에 생성할 토큰의최대를 개수 지정.
- `model_name` : 모델을 지정한다.


```python
from langchain_openai import ChatOpenAI

gpt = ChatOpenAI(model="gpt-4o-mini", temperature=0)

print(gpt.invoke("사랑이 뭔가요?").content)
```

```python
gpt = ChatOpenAI(model="gpt-4o-mini", temperature=0.9)

print(gpt.invoke("사랑이 뭔가요?").content)
```

```python
answer = gpt.stream("사랑이 뭔가요?")
for token in answer:
    print(token.content, end="", flush=True)
```

# Anthropic
- 미국회사이며, OpenAI직원들이 나와서 만든 모델.
- 원래 비용이 문제였는데, 많이 다운됌!


```python
from langchain_anthropic import ChatAnthropic
anthropic = ChatAnthropic(model="claude-3-5-sonnet-20240620")

answer = anthropic.stream("사랑이 뭔가요?")

for s in answer:
    print(s)
```

# Cohere
